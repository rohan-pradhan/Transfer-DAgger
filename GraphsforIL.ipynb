## Original

```
import matplotlib.pyplot as plt

x = [13.82228, 72.8198, 79.0250, 83.1105]

y_old_track = [819.2999, 815.7999, 23.7127, 21.457]

y_new_track = [312.219, 806.999, 44.394, 16.862]
y_original = [819.499]*4

y_original_new = [809.3999]*4

plt.figure(figsize=(12,8))



plt.grid('True')
plt.plot(x, y_old_track, 'bo-', label="Original Track")
plt.plot(x, y_original, 'b--', label="Original Track with Original Color")
plt.plot(x, y_original_new, 'r--', label="New Unseen Track with Original Color")
plt.plot(x, y_new_track, 'ro-', label="New Unseen Track")
# plt.plot(mode, 'b', label="Mode")
# plt.plot(x_adv, 'r--', label = "No Manipulation")
plt.ylabel('Cumulative Reward')
plt.xlabel('Similarity in Road and Grass Color ')
plt.title('Similarity in Road and Grass Color vs. Cumulative Reward Obtained When Using Original DAgger Policy')
plt.legend()
plt.show()
```
